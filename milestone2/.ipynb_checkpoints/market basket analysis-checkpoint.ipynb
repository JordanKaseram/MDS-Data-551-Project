```python
import pandas as pd


df_raw = pd.read_csv("data/SuperStoreOrders.csv")
df = df_raw.copy()
df["sales"] = (
    df["sales"]
        .astype(str)
        .str.replace(",", "", regex=False)
        .str.strip()
)

df["sales"] = pd.to_numeric(df["sales"], errors="coerce")
df["margin"] = df["profit"] / df["sales"]

df["order_date"] = pd.to_datetime(df["order_date"], format="mixed", dayfirst=True, errors="coerce")
df["ship_date"]  = pd.to_datetime(df["ship_date"],  format="mixed", dayfirst=True, errors="coerce")
df["shipping_delay"] = (df["ship_date"] - df["order_date"]).dt.days
```

```python
import networkx as nx
import matplotlib.pyplot as plt
import numpy as np

top_edges = rules.head(40).copy()  # tune 20~80 depending on readability

G = nx.DiGraph()
for _, r in top_edges.iterrows():
    G.add_edge(r["prod1"], r["prod2"], weight=float(r["lift"]), support=float(r["support"]))

plt.figure(figsize=(10, 8))
pos = nx.spring_layout(G, seed=7, k=0.7)

# Edge widths by lift
weights = [G[u][v]["weight"] for u, v in G.edges()]
w = np.array(weights)
edge_widths = 1 + 3*(w - w.min())/(w.max() - w.min() + 1e-9)

nx.draw_networkx_nodes(G, pos, node_size=900)
nx.draw_networkx_labels(G, pos, font_size=9)
nx.draw_networkx_edges(G, pos, width=edge_widths, arrows=True, alpha=0.5)

plt.title("MBA Network (edges weighted by Lift)")
plt.axis("off")
plt.show()
```

# 0) Prep: season + core aggregates

```python
import pandas as pd
import numpy as np
import altair as alt

alt.data_transformers.disable_max_rows()

# --- Season column (4 full seasons) ---

df_f = df.copy()

df_f["month"] = df_f["order_date"].dt.month

df_f["season"] = np.select(
    [
        df_f["month"].isin([3,4,5]),      # Spring
        df_f["month"].isin([6,7,8]),      # Summer
        df_f["month"].isin([9,10,11]),    # Fall
        df_f["month"].isin([12,1,2])      # Winter
    ],
    ["Spring", "Summer", "Fall", "Winter"],
    default="Unknown"
)

# Optional: make it ordered (nice for dashboard sorting)
df_f["season"] = pd.Categorical(
    df_f["season"],
    categories=["Spring","Summer","Fall","Winter"],
    ordered=True
)

# Example filters like your sketch (period=season, segment, year)
# df_f = df_f[df_f["season"].isin(["Summer"])]
# df_f = df_f[df_f["segment"].isin(["Consumer"])]
# df_f = df_f[df_f["year"].isin([2024])]
```

# 1) KPI strip (small cards)

```python
# -----------------------------
# KPI calculations
# -----------------------------

total_sales = df_f["sales"].sum()
total_profit = df_f["profit"].sum()

avg_margin = (total_profit / total_sales * 100) if total_sales != 0 else np.nan
avg_discount = df_f["discount"].mean() * 100 if "discount" in df_f.columns else np.nan

# Customers
if "customer_name" in df_f.columns:
    n_customers = df_f["customer_name"].nunique()
else:
    n_customers = df_f["order_id"].nunique()

# Orders
n_orders = df_f["order_id"].nunique()

# NEW KPI 1: Purchase Frequency (orders per customer)
purchase_frequency = (n_orders / n_customers) if n_customers != 0 else np.nan

# NEW KPI 2: Avg Products per Transaction
total_quantity = df_f["quantity"].sum()
avg_products_per_txn = (total_quantity / n_orders) if n_orders != 0 else np.nan


# -----------------------------
# KPI row
# -----------------------------

kpi_row = alt.hconcat(
    kpi_value("Total Customers", f"{n_customers:,}"),
    kpi_value("Revenue", f"${total_sales:,.0f}"),
    kpi_value("Profit", f"${total_profit:,.0f}"),
    kpi_value("Margin %", f"{avg_margin:.1f}%"),
    kpi_value("Discount %", f"{avg_discount:.1f}%"),
    kpi_value("Frequency", f"{ :.1f}"),
    kpi_value("Avg Products / Txn", f"{avg_products_per_txn:.1f}"),
    spacing=18
).properties(
    title="Campaign Optimize Profit Strategy"
).configure_title(
    anchor="start",
    fontSize=18
)

kpi_row
```

# 2) Middle-left: Subcategory discovery bubble (your main graph)

```python
total_orders = df_f["order_id"].nunique()

subcat = (
    df_f.groupby(["category", "sub_category"], as_index=False)
        .agg(
            orders=("order_id", "nunique"),
            sales=("sales", "sum"),
            profit=("profit", "sum"),
        )
)
subcat["freq_pct"] = np.where(total_orders>0, subcat["orders"]/total_orders*100, 0.0)
subcat["margin_pct"] = np.where(subcat["sales"]!=0, subcat["profit"]/subcat["sales"]*100, np.nan)
subcat = subcat.replace([np.inf, -np.inf], np.nan).dropna(subset=["margin_pct"])

x_med = float(subcat["freq_pct"].median())
y_med = float(subcat["margin_pct"].median())

bubble = (
    alt.Chart(subcat)
    .mark_circle(opacity=0.75, stroke="white", strokeWidth=0.5)
    .encode(
        x=alt.X("freq_pct:Q", title="frequency of buying (% of orders)"),
        y=alt.Y("margin_pct:Q", title="% margin"),
        size=alt.Size("sales:Q", scale=alt.Scale(range=[40, 1200]), title="sales"),
        color=alt.Color("category:N", title="category"),
        tooltip=[
            "sub_category:N","category:N",
            alt.Tooltip("freq_pct:Q", format=".2f", title="freq %"),
            alt.Tooltip("margin_pct:Q", format=".2f", title="margin %"),
            alt.Tooltip("sales:Q", format=",.0f", title="sales"),
        ],
    )
    .properties(width=640, height=340, title="Subcategory discovery")
)

labels = alt.Chart(subcat).mark_text(
    align="left", dx=7, dy=-7, fontSize=10, color="#2B2B2B"
).encode(
    x="freq_pct:Q", y="margin_pct:Q", text="sub_category:N"
)

vline = alt.Chart(pd.DataFrame({"x":[x_med]})).mark_rule(strokeDash=[6,6], opacity=0.5).encode(x="x:Q")
hline = alt.Chart(pd.DataFrame({"y":[y_med]})).mark_rule(strokeDash=[6,6], opacity=0.5).encode(y="y:Q")

bubble_panel = (bubble + labels + vline + hline)
bubble_panel
```

# 3) Middle-right: Bundle subcategory strategy (MBA table)

# 4) Bottom-left: Top 5 products bar-table (FIXED to look like one table)

```python
# Build top5 products (you can later filter to a selected subcategory)
prod = (
    df_f.groupby(["sub_category","product_name"], as_index=False)
        .agg(
            sales=("sales","sum"),
            profit=("profit","sum"),
            customers=("customer_name","nunique") if "customer_name" in df_f.columns else ("order_id","nunique"),
        )
)
prod["margin_pct"] = np.where(prod["sales"]!=0, prod["profit"]/prod["sales"]*100, np.nan)
prod = prod.replace([np.inf,-np.inf], np.nan).dropna(subset=["margin_pct"])

top5 = prod.sort_values("margin_pct", ascending=False).head(5).copy()
top5["label"] = top5["sub_category"] + " • " + top5["product_name"]

y = alt.Y("label:N", sort=alt.SortField(field="margin_pct", order="descending"), title=None,
          axis=alt.Axis(labelLimit=360))

label_col = alt.Chart(top5).mark_text(align="left", baseline="middle").encode(
    y=y,
    text="label:N"
).properties(width=360, height=220, title="Breakdown: Top 5 products (sorted by % margin)")

# Bars: hide y axis here (so you don't repeat labels)
no_y = alt.Y("label:N", sort=alt.SortField(field="margin_pct", order="descending"), title=None, axis=None)

m_bar = alt.Chart(top5).mark_bar(cornerRadiusEnd=3).encode(
    y=no_y,
    x=alt.X("margin_pct:Q", title="% margin"),
).properties(width=170, height=220)

s_bar = alt.Chart(top5).mark_bar(cornerRadiusEnd=3).encode(
    y=no_y,
    x=alt.X("sales:Q", title="sales"),
).properties(width=170, height=220)

c_bar = alt.Chart(top5).mark_bar(cornerRadiusEnd=3).encode(
    y=no_y,
    x=alt.X("customers:Q", title="# customers"),
).properties(width=170, height=220)

bar_table_panel = alt.hconcat(label_col, m_bar, s_bar, c_bar, spacing=12).resolve_scale(y="shared")
bar_table_panel
```

# 5) Bottom-right: Campaign framework — Margin vs Discount by Subcategory

```python
bins = np.arange(0, 0.41, 0.05)
df_tmp = df_f.copy()
df_tmp["discount"] = pd.to_numeric(df_tmp["discount"], errors="coerce")

df_tmp["disc_bin"] = pd.cut(df_tmp["discount"], bins=bins, include_lowest=True)

disc = (
    df_tmp.groupby(["sub_category","disc_bin"], as_index=False)
          .agg(sales=("sales","sum"), profit=("profit","sum"))
)

disc["margin_pct"] = np.where(disc["sales"]!=0, disc["profit"]/disc["sales"]*100, np.nan)
disc = disc.dropna(subset=["margin_pct","disc_bin"])

# numeric midpoint (SAFE)
disc["disc_mid"] = disc["disc_bin"].map(lambda x: float((x.left + x.right)/2) * 100)

# ✅ IMPORTANT: remove Interval column OR convert it
disc = disc.drop(columns=["disc_bin"])

discount_panel = (
    alt.Chart(disc)
    .mark_line(point=True)
    .encode(
        x=alt.X("disc_mid:Q", title="% discount"),
        y=alt.Y("margin_pct:Q", title="% margin"),
        color=alt.Color("sub_category:N", title="breakdown by subcat"),
        tooltip=[
            "sub_category:N",
            alt.Tooltip("disc_mid:Q", format=".0f", title="discount %"),
            alt.Tooltip("margin_pct:Q", format=".2f", title="margin %")
        ],
    )
    .properties(width=420, height=220, title="Optimize discount (breakdown by subcat)")
)
discount_panel
```

```python
# =========================
# Discount Guardrail Heatmap (Altair) — POLISHED VERSION
# Upgrades:
# 1) Strategy bins: 0–5, 5–10, 10–15, 15–20, 20–25, 25–35, >35
# 2) Stable cells only (min orders/sales)
# 3) Diverging color centered at 0 + CLAMP domain to avoid outliers dominating
# 4) Sort subcategories by sensitivity at >35% (worst first)
# =========================

import numpy as np
import pandas as pd
import altair as alt

alt.data_transformers.disable_max_rows()

df_f = df.copy()
df_f["discount"] = pd.to_numeric(df_f["discount"], errors="coerce")

# ---- 1) bins ----
bins   = [0, 0.05, 0.10, 0.15, 0.20, 0.25, 0.35, 1.00]
labels = ["0–5%", "5–10%", "10–15%", "15–20%", "20–25%", "25–35%", ">35%"]

df_f["disc_bin"] = pd.cut(df_f["discount"], bins=bins, labels=labels, include_lowest=True)

# ---- 2) aggregate ----
disc = (
    df_f.groupby(["sub_category", "disc_bin"], as_index=False)
        .agg(
            sales=("sales", "sum"),
            profit=("profit", "sum"),
            orders=("order_id", "nunique"),
        )
)

disc["margin_pct"] = np.where(disc["sales"] != 0, disc["profit"] / disc["sales"] * 100, np.nan)
disc = disc.dropna(subset=["margin_pct", "disc_bin"])

# ---- 3) stability filter (tune as needed) ----
MIN_ORDERS = 20
MIN_SALES  = 3000
disc = disc[(disc["orders"] >= MIN_ORDERS) & (disc["sales"] >= MIN_SALES)].copy()

# ---- 4) keep Top subcats by sales (avoid clutter) ----
TOP_N = 15
top_subcats = (
    df_f.groupby("sub_category")["sales"].sum()
        .sort_values(ascending=False)
        .head(TOP_N)
        .index
)
disc_top = disc[disc["sub_category"].isin(top_subcats)].copy()

# Altair-safe (no category/interval issues)
disc_top["disc_bin"] = disc_top["disc_bin"].astype(str)

# ---- 5) sort rows by sensitivity in >35% bucket (worst margin first) ----
sens = disc_top[disc_top["disc_bin"] == ">35%"][["sub_category", "margin_pct"]].copy()

if len(sens) > 0:
    row_order = sens.sort_values("margin_pct")["sub_category"].tolist()
else:
    # fallback: sort by total sales desc if >35% has no data
    row_order = (
        df_f.groupby("sub_category")["sales"].sum()
            .loc[top_subcats]
            .sort_values(ascending=False)
            .index
            .tolist()
    )

# ---- 6) polished heatmap ----
# CLAMP domain so one extreme (e.g. -80%) doesn't flatten everything
CLAMP = 40  # show -40% to +40% clearly
disc_top["margin_clamped"] = disc_top["margin_pct"].clip(-CLAMP, CLAMP)

heatmap = (
    alt.Chart(disc_top)
    .mark_rect()
    .encode(
        x=alt.X("disc_bin:N", sort=labels, title="% discount (strategy bucket)"),
        y=alt.Y("sub_category:N", sort=row_order, title=None),
        color=alt.Color(
            "margin_clamped:Q",
            title="Margin %",
            scale=alt.Scale(scheme="redblue", domain=[-CLAMP, CLAMP], domainMid=0)
        ),
        tooltip=[
            alt.Tooltip("sub_category:N", title="Subcategory"),
            alt.Tooltip("disc_bin:N", title="Discount bucket"),
            alt.Tooltip("orders:Q", format=",.0f", title="# Orders"),
            alt.Tooltip("sales:Q", format=",.0f", title="Sales"),
            alt.Tooltip("profit:Q", format=",.0f", title="Profit"),
            alt.Tooltip("margin_pct:Q", format=".2f", title="Margin % (raw)"),
        ]
    )
    .properties(width=820, height=420, title="Discount Guardrail Framework (Top subcategories)")
    .configure_title(anchor="start", fontSize=16)
    .configure_axis(labelFontSize=11, titleFontSize=12)
)

heatmap
```
